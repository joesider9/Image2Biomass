```python
import os
import copy
import joblib
```

```python
from eforecast.init.initialize import initializer
from eforecast.datasets.dataset_creator import DatasetCreator
from configuration.config import config
static_data = initializer(config())
```

```python
dataset = DatasetCreator(static_data, recreate=False, train=True)
dataset.create_target()
dataset.create_row_datasets()
dataset.create_image_dataset(parallel=True)
```

```python
from eforecast.datasets.data_preprocessing.data_pipeline import DataPipeline
pipeline = DataPipeline(static_data, recreate=True, online=False, train=True)
pipeline.fit_pipe()
```

```python
from eforecast.clustering.clustering_manager import ClusterOrganizer
from eforecast.datasets.data_preprocessing.data_split import Splitter
splitter = Splitter(static_data, is_online=False, train=True)
splitter.split(refit=True)
cluster_organizer = ClusterOrganizer(static_data, is_online=False, train=True, refit=True)
cluster_organizer.fit()
cluster_organizer.create_clusters_and_cvs()
```

```python
from eforecast.feature_selection.feature_selection_fit import FeatureSelector
feature_selector = FeatureSelector(static_data, recreate=False, online=False, train=True)
feature_selector.fit(recreate_lstm=False)
feature_selector.create_backup()
```

```python
from eforecast.deep_models.pytorch_2x.network import DeepNetwork

cluster_name = 'global'
method = 'TIMM'
cluster_dir = os.path.join(static_data['path_model'], cluster_name)
exp_timm = {
    'images': [('vit_net', 1),
               ('Flatten', []),
               ('dense', 1024),
               ('dense', 256),
               ],
    'row_stats': [('dense', {128}),
                         ],
    'output': [('cross_attention', {1}),
               ('dense', 64),
               ]
}
data_types = {
    'images': {},
    'row_stats': {'scale_row_method': 'minmax',
                'feature_selection_method': None},
}
trial_number = 0
params = copy.deepcopy(static_data['TIMM'])
experiment_params = {'trial_number': trial_number,
                 'method': method,
                 'name': cluster_name,
                 'cluster_dir': cluster_dir,
                 'cluster': {'cluster_name':cluster_name,
                             'cluster_path':cluster_dir},
                 'data_types': data_types,
                 'image_in_data': True,
                 'experiment_tag': 'timm_net5',
                 'experiment': exp_timm}
params.update(experiment_params)
path_weights = os.path.join(cluster_dir, method, f'test_{trial_number}')
os.makedirs(path_weights, exist_ok=True)
params['model_name'] = params['model_name'][trial_number]
params['max_iterations'] = 800

cv_masks = joblib.load(os.path.join(cluster_dir, 'cv_mask.pickle'))
model = DeepNetwork(static_data, path_weights, params, train=True, refit=True)
model.fit(cv_masks, gpu_id=0)
```

```python
clusters = joblib.load(os.path.join(static_data['path_model'], 'clusters.pickle'))
global_name = 'global'
method = 'TIMM'
exp_timm = {
    'images': [('vit_net', 1),
               ('Flatten', []),
               ('dense', 1024),
               ('dense', 256),
               ],
    'row_stats': [('dense', {128}),
                         ],
    'output': [('cross_attention', {1}),
               ('dense', 64),
               ]
}
data_types = {
    'images': {},
    'row_stats': {'scale_row_method': 'minmax',
                'feature_selection_method': None},
}
global_dir = os.path.join(static_data['path_model'], global_name)
for cluster_name, cluster_dir in clusters.items():
    trial_number = 0
    tl_path_weights = os.path.join(global_dir, method, f'test_{trial_number}')
    params = copy.deepcopy(static_data['TIMM'])
    experiment_params = {'trial_number': trial_number,
                     'method': method,
                     'name': cluster_name,
                     'cluster_dir': cluster_dir,
                     'cluster': {'cluster_name':cluster_name,
                                 'cluster_path':cluster_dir},
                     'tl_path_weights': tl_path_weights,
                     'data_types': data_types,
                     'image_in_data': True,
                     'experiment_tag': 'timm_net5',
                     'experiment': exp_timm}
    params.update(experiment_params)
    path_weights = os.path.join(cluster_dir, method, f'test_{trial_number}')
    os.makedirs(path_weights, exist_ok=True)
    params['model_name'] = params['model_name'][trial_number]
    params['max_iterations'] = 200
    # params['learning_rate'] = 0.5e-4
    cv_masks = joblib.load(os.path.join(cluster_dir, 'cv_mask.pickle'))
    model = DeepNetwork(static_data, path_weights, params, train=True, refit=True)
    model.fit(cv_masks, gpu_id=0)
```

```python
from eforecast.combine_predictions.combine_predictions_fit import CombinerFit
from eforecast.prediction.predict import Predictor
from eforecast.prediction.evaluate import Evaluator
predictor = Predictor(static_data, train=True)
predictor.predict_regressors(average=True, parallel=False)
combiner = CombinerFit(static_data, refit=True)
combiner.fit_methods()
predictor = Predictor(static_data, train=True)
predictor.predict_combine_methods()
predictor.compute_predictions_averages(only_methods=False, only_combine_methods=False)
evaluator = Evaluator(static_data, refit=True)
evaluator.evaluate_averages()
```
